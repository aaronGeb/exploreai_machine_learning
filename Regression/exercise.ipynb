<li>Fit a simple linear regression model using sklearn's LinearRegression.
<li>Predict the agricultural yield for a year with 1150 mm of annual rainfall.

```python
annual_rainfall = [800, 1200, 1000, 1500, 1100, 1300, 900, 1400, 950, 1250]
agricultural_yield = [3.2, 4.8, 4.0, 5.5, 4.2, 5.0, 3.5, 5.3, 3.8, 4.9]
annual_rainfall = np.array(annual_rainfall)
agricultural_yield = np.array(agricultural_yield)
```

```python
model= LinearRegression()
```

```python
model.fit(annual_rainfall.reshape(-1,1), agricultural_yield)
```

```python
predicted_yield = model.predict([[1100]])
predicted_yield
```
